```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

cars=pd.read_csv(r'C:\Users\Jalal\Desktop\Programing\wkdir\git7\INTRO-TO-DATA-ANALYSIS\DATA\cars_multivariate.csv',na_values=['?'])
cars.head()
cars['model']=cars.model.astype('object')
cars['origin']=cars.origin.astype('object')
cars.horsepower.astype('float')
cars=cars[cars.horsepower.notna()]
numeric_vars = ['mpg','displacement','horsepower','weight','acceleration']
categorical_vars = ['model', 'origin', 'cylinders','car_name']
```
